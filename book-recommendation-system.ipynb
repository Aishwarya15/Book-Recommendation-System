```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv('../input/books.csv')
```

```python
#display books dataset
books
```

```python
#check if any bookids are NULL (missing/wrong data)
books['book_id'].isnull().sum()
```

```python
ratings = pd.read_csv('../input/ratings.csv')
```

```python
#display ratings dataset
ratings
```

```python
#select the important columns from the books dataset
books_dataset = pd.DataFrame(books, columns=['book_id', 'authors', 'title', 'average_rating'])
```

```python
#sort according to bookid
books_dataset = books_dataset.sort_values('book_id')
```

```python
#merge books dataframe and ratings dataset
books_data = pd.merge(books_dataset, ratings, on='book_id')
```

```python
#display this merged dataset
books_data
```

```python
#create a table which displays user and ratings for each book title 
each_book_rating = pd.pivot_table(books_data, index='user_id', values='rating', columns='title', fill_value=0)
```

```python
#display this table
each_book_rating
```

```python
#find correlation coefficient for each book
book_corr = np.corrcoef(each_book_rating.T)
```

```python
#2d array converted into list because we require indexing further
book_list=  list(each_book_rating)
book_titles =[] 
for i in range(len(book_list)):
    book_titles.append(book_list[i])
```

```python
#display book_titles 
book_titles
```

```python
#select index of given book from the list of books
book = 'The Alchemist'
book_index = book_titles.index(book)
```

```python
#find out the correlation coefficient of given book through its index
corr_score = book_corr[book_index]
```

```python
#display corr coeff in descending order
print(sorted(corr_score, reverse=True))
```

```python
#create a condition for identifying similar books
condition = (corr_score >= 0.1)
```

```python
#get the book titles that satisfy the condition
np.extract(condition, book_titles)
```

```python
#function for getting recommendations for a list of favourite books
def get_recommendation(books_list):
    book_similarities = np.zeros(book_corr.shape[0])
    
    for book in books_list:    

        book_index = book_titles.index(book)
        book_similarities = book_corr[book_index] 
        #print(book_similarities)
        
    book_preferences = []
    for i in range(len(book_titles)):
        book_preferences.append((book_titles[i],book_similarities[i]))
    #print(book_preferences)    
    return sorted(book_preferences, key= lambda x: x[1], reverse=True)
```

```python
#add a few books to the favourites list
my_fav_books = ['The Alchemist','The Adventures of Sherlock Holmes','The Great Gatsby','To Kill a Mockingbird','The Da Vinci Code (Robert Langdon, #2)','The Fellowship of the Ring (The Lord of the Rings, #1)']
```

```python
#apply function to the favourites list
book_recommendations = get_recommendation(my_fav_books)
```

```python
#print the answer given by the function
print('Other books you may like :')
print('-'*26)
i=0
c=0
while c < 9:
    book_to_read = book_recommendations[i][0]
    i += 1
    if book_to_read in my_fav_books:
        continue
    else:
        print(book_to_read)
        c += 1
```
